```python
import requests
import pandas as pd
from pathlib import Path
from bs4 import BeautifulSoup
```

```python
web_path = "https://unicode.org/emoji/charts/full-emoji-list.html"
web_content = requests.get(web_path).content
```

```python
web_parse = BeautifulSoup(web_content, 'html')
table_emojis = web_parse.find("table")
web_tbody = table_emojis.find("tbody")
```

```python
with open('web_emojis_data.csv', 'w') as emo:
    emo.write('number;code;chars;name\n')
    for tr in table_emojis.find_all('tr'):
        if len(tr.find_all('td')) > 1:
            line = []
            for td in tr.find_all('td'):
                value = td.text.strip()
                if 'rchars' in td['class']: line.append(value)
                if 'code' in td['class']: line.append(value)
                if 'chars' in td['class']: line.append(value)
                if 'name' in td['class']: line.append(value)
            c_write = ";".join(line)+"\n"
            emo.write(c_write)
```

```python
import json
data_json = {}
with open('web_emojis_data.json', 'w') as emo:
    for tr in table_emojis.find_all('tr'):
        if len(tr.find_all('td')) > 1:
            line = {}
            for td in tr.find_all('td'):
                value = td.text.strip()
                if 'rchars' in td['class']: line['rchars'] = value
                if 'code' in td['class']: line['code'] = value
                if 'chars' in td['class']: line['chars'] = value
                if 'name' in td['class']: line['name'] = value
            data_json[line['chars']] = line

    json.dump(data_json, emo, indent=4)
```
